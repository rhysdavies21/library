```
# Import libraries
import pandas as pd
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib import style
import pandas_datareader as web
from yahoofinancials import YahooFinancials

# Scrape data of FTSE constituents from Wikipedia
table=pd.read_html('https://en.wikipedia.org/wiki/FTSE_100_Index')

# Build dataframe
df = table[2]
print('Check shape: ', df.shape)
df.head(5)

# Ref https://pypi.org/project/yahoofinancials/
```

```
# Build list of importaed dividend yield from Yahoo
# This can take time to import data
yahoo_ticker = []
div_yield = []

for ticker in list(df['Ticker']):
  
  if ticker[-1:] == ".": 
    ticker_i = ticker + 'L'
  elif ticker[-2:][0] == ".":
    ticker_i = ticker
  else: 
    ticker_i = ticker + '.L'
  
  yahoo_ticker.append(ticker_i)

  try:
    div_yield.append("{:.2%}".format(YahooFinancials(ticker_i).get_summary_data()[ticker_i]['dividendYield']))
  except TypeError:
    div_yield.append('Error')
```

```
# Add data into dataframe
df['Yahoo_Ticker'] = yahoo_ticker
df['Div_Yield'] = div_yield

print('Shape of dataframe:', df.shape)
df.head(5)
```

```
# Drop names without Div_Yield
df_clean = df[df.Div_Yield != 'Error']

# Sort by Div_Yield
df_clean.sort_values(by=['Div_Yield'], ascending=False)
```
